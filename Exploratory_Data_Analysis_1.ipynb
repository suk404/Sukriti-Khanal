```r
ggplot(data=mtcars)+geom_point(mapping=aes(x=mpg,y=hp,color=factor(cyl)))
```

## **Plot types**

```r
df<- data.frame(
  name=c("Male","Female","Others"),
  count= c(30,35,5)
)
```

```r
df
```

```r
# if the values are already provided, we use identity in the statistics.
ggplot(data=df,aes(x=name,y=count))+geom_bar(stat='identity')
```

```r
ggplot(data=df,aes(x=name,y=count))+geom_bar(stat='identity',color="red",fill="steelblue",width = 0.5)
```

```r
ggplot(data=df,aes(x=name,y=count))+geom_bar(stat='identity',color="red",fill="steelblue",width = 0.5)+coord_flip()+theme_minimal()
```

```r
p<- ggplot(data=df,aes(x=name,y=count))+geom_bar(stat='identity',color="red",fill="steelblue",width = 0.5)
p+geom_text(aes(label=count),vjust=-1)
# vjust means the vertical adjust and -1 means raise it 1 unit above the bar.
```
