```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url="/content/Dataset Car_Sukriti.xlsx"
```

```python
file1=pd.read_excel(url)
```

```python
file1.head()
```

```python
import scipy.stats
```

```python
file1.info()
```

```python
Y=file1['Price']
```

```python
X=file1.drop('Price',axis=1)
```

```python
X.columns
```

```python
print(X)
```

```python
X.columns = X.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
fitted_meodel= clf.fit(x_train,y_train)
```

```python
pred_y=clf.predict(x_test)
```

```python
print("Predictions:", pred_y[1:5])
print("Labels:", list(y_test[1:5]))
```

```python
clf.score(x_test,y_test)
```

```python
len(pred_y)
```

```python
type(pred_y)
```

```python
new_data1=pd.DataFrame()
```

```python
new_data1['Actual_Price']=y_test
```

```python
new_data1['Predicted Price'] = pred_y
```

```python
new_data1.head()
```

```python
# Let’s measure this regression model’s RMSE on the whole training set
from sklearn.metrics import mean_squared_error


lin_rmse = mean_squared_error(y_test, pred_y, squared=False)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, pred_y)
mae
```

```python
new_data1.to_csv('predicted_prices.csv', index=False)
```
