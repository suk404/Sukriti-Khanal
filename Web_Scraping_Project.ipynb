```python
def get_html_doc(url):
  response=requests.get(url)
  return(response.text)
```

```python
def parse_html(html_text):
  soup=BeautifulSoup(html_text,"html.parser")
  return soup
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
master_data=[]
for ing in range(1,5):
  url="https://www.sastodeal.com/electronic/mobile.html?p={}".format(ing)

  html_text=get_html_doc(url)
  soup=parse_html(html_text)
  master_data.extend(get_data(soup))
```
