**Contours**

```python
bw=cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,-3)
```

```python
cv2_imshow(bw)
```

```python
bw
```

```python
inv_bw=cv2.bitwise_not(bw)
cv2_imshow(inv_bw)
```

```python
bw_and=cv2.bitwise_and(bw,inv_bw)
cv2_imshow(bw_and)
```

```python
bw_or=cv2.bitwise_or(bw,inv_bw)
cv2_imshow(bw_or)
#true or false means true and true is 255 which means white pixels.
# bw=cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,15,-3)
```

**Contours- will help you separate black pixels from white pixels**

```python
cv2_imshow(bw)
```

```python
contours,hierarchy=cv2.findContours(bw,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(image, contours,-1,(0,0,255),1) #-1 draws all the contours you can also select which contour you want to draw
#3 or 1 could be used at the end above which represents the thickness of the line.
cv2_imshow(image)
```
