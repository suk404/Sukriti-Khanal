```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_csv('/content/seabornli.csv')
```

```python
relation=data_df.corr()
```

```python
data_df.columns
```

```python
relation
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=relation,cmap="Oranges")
plt.title("Correlation of features")
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
plt.show()
```
