**Bar plot**

```r
ggplot(data=data,aes(x=total_bedrooms,fill=factor(median_house_value)))+geom_bar(stat='count')+theme_minimal()
```

```r
p1<- ggplot(data=data,aes(x=total_rooms))+geom_histogram(bins=sqrt(length(data$total_rooms)),fill='tomato',color='white')+theme_minimal()
p1
```

```r
p1<- ggplot(data=data,aes(x=total_bedrooms))+geom_histogram(bins=sqrt(length(data$total_bedrooms)),fill='tomato',color='white')+theme_minimal()
p1
```

```r
p2<- p1+geom_vline(xintercept=mean(data$total_bedrooms, color='red',linetype="dashed",size=1))
p2
```

```r
p1<- ggplot(data=data,aes(x=median_house_value))+geom_histogram(bins=sqrt(length(data$median_house_value)),fill='tomato',color='white')+theme_minimal()
p1
```

```r
ggplot(data=data,aes(x=total_bedrooms,y=median_house_value,fill=factor(median_house_value)))+geom_bar(stat='identity')+theme_minimal()
```
